```python
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import pandas as pd
```

```python
# Utilities: Home --> Amenities --> Scrape Community Amenities --> Scrape Apartment Amenities

#box 1: section class= "pp-wrapper pp-container"

#box 2: section class= "pp-content full-width"

# ---> community: <div class="amenity-group ca">

# ---> apartment: <div class="amenity-group aa">
```

```python
PATH = Service("/Users/bellabell/Downloads/Bing Spring 2023/Coding 2 /Coding 2 Labs/chromedriver.exe")
driver = webdriver.Chrome(service=PATH)

driver.get("https://www.theprintinghouseny.com") #opens website home screen

csv_file = open("printing_house_amenities.csv", "w", newline="", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Community Amenities", "Apartment Amenities"])

link = driver.find_element(By.LINK_TEXT, "Amenities") # allows us to type text related to link & access element from it
link.click() #clicks link

try: 
    for comm_amenities in driver.find_elements(By.XPATH, '/html/body/div[1]/div[2]/section/section/div[1]'):
        print(comm_amenities.text)
        comm = comm_amenities.text.split("\n")
        
    
    for apart_amenities in driver.find_elements(By.XPATH, '/html/body/div[1]/div[2]/section/section/div[2]'):
        print(apart_amenities.text)
        apart = apart_amenities.text.split("\n")
    
    
    
    for index, amenity in enumerate(comm):
        community_amenity = amenity
        try:
            apartment_amenity = apart[index]
        except:
            apartment_amenity = ""
            
        csv_writer.writerow([community_amenity, apartment_amenity])
    csv_file.close()
        
    
finally:
    driver.quit()
```
