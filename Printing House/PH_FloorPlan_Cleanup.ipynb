```python
import pandas as pd

PH_Floors = pd.read_csv("printinghouse_floor_plansTEST.csv")
PH_Floors
```

```python
PH_Floors2 = PH_Floors.copy() #make a copy
PH_Floors2 = PH_Floors2.dropna() #drop all other rows with an NaN
PH_Floors2
```

```python
#convert all columns into str

PH_Floors2["RoomType"] = PH_Floors2["RoomType"].astype(str) 
PH_Floors2["Rent"] = PH_Floors2["Rent"].astype(str) 
PH_Floors2["BedsBaths"] = PH_Floors2["BedsBaths"].astype(str)
PH_Floors2["Sqft"] = PH_Floors2["Sqft"].astype(str)

PH_Floors2
```

```python
PH_FloorsC = PH_Floors2.copy()

PH_FloorsC.Rent = PH_FloorsC.Rent.str[7:14] 
PH_FloorsC.Rent

PH_FloorsC
```

```python
PH_FloorsC.RoomType = PH_FloorsC.RoomType.str[13:25]  
PH_FloorsC.RoomType

PH_FloorsC
```

```python
PH_FloorsC.BedsBaths = PH_FloorsC.BedsBaths.str[13:28] 
PH_FloorsC.BedsBaths

PH_FloorsC
```

```python
PH_FloorsC.Sqft = PH_FloorsC.Sqft.str[11:25] 
PH_FloorsC.Sqft

PH_FloorsC
```
